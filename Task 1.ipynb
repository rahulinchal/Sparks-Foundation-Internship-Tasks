### Defining dependent and independent variables

```python
#x = df[['Hours']].values
#y = df[['Scores']].values

x = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
```

### Splitting train and test 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

### Importing Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

### Fitting the model 

```python
lr.fit(x_train, y_train)
```

### Predictind the values 

```python
lr_pred = lr.predict(x_test)
lr_pred
```

```python
y_test
```

### Comparing the model

```python
# Comparing Actual vs Predicted
df_new = pd.DataFrame({'Actual': y_test, 'Predicted': lr_pred}) 
df_new
```

### Printing the scores

```python
print(lr.score(x_train, y_train)*100)
print(lr.score(x_test, y_test)*100)
```

### Evaluating the Performance

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import math
from sklearn.metrics import r2_score
```

```python
print(f'Mean Absolute Error = {mean_absolute_error(y_test, lr_pred)}')
print(f'Mean Squared Error = {mean_squared_error(y_test, lr_pred)}')
print(f'Root Mean Squared Error = {math.sqrt(mean_squared_error(y_test, lr_pred))}')
print(f'r2 = {r2_score(y_test, lr_pred)}')
```
